The block of code below is used to show the images all at once. plt.subplot(1,3,1) means do a subplot of the  first image on row number 1 and column number 3

```python
plt.figure(figsize=(10,6))
plt.subplot(1,3,1)
plt.imshow(r,cmap='Reds')
plt.subplot(1,3,2)
plt.imshow(g,cmap='Greens')
plt.subplot(1,3,3)
plt.imshow(b,cmap='Blues')
plt.show()
plt.imshow(img1)
```
